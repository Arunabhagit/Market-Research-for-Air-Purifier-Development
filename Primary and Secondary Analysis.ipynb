List out top 2 and bottom 2 prominent pollutants for each state of southern India.
(Consider data post covid: 2022 onwards)

```python
import pandas as pd




aqi['date'] = pd.to_datetime(aqi['date'], dayfirst=True)


aqi_post_covid = aqi[aqi['date'].dt.year >= 2022]


southern_states = ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Telangana']


south_df = aqi_post_covid[aqi_post_covid['state'].isin(southern_states)]


pollutant_counts = south_df.groupby(['state', 'prominent_pollutants']).size().reset_index(name='count')


def get_top_bottom(df):
    return pd.concat([df.nlargest(2, 'count'), df.nsmallest(2, 'count')])

result = pollutant_counts.groupby('state').apply(get_top_bottom).reset_index(drop=True)


print(result)
```

Does AQI improve on weekends vs weekdays in Indian metro cities (Delhi,
Mumbai, Chennai, Kolkata, Bengaluru, Hyderabad, Ahmedabad, Pune)?
(Consider data from last 1 year)

```python



aqi['date'] = pd.to_datetime(aqi['date'], dayfirst=True)


latest_date = aqi['date'].max()
one_year_ago = latest_date - pd.DateOffset(years=1)
last_year_df = aqi[(aqi['date'] >= one_year_ago) & (aqi['date'] <= latest_date)]


metro_cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bengaluru', 'Hyderabad', 'Ahmedabad', 'Pune']


metro_df = last_year_df[last_year_df['area'].isin(metro_cities)]

metro_df['day_type'] = metro_df['date'].dt.dayofweek.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')


avg_aqi_comparison = metro_df.groupby(['area', 'day_type'])['aqi_value'].mean().reset_index()


print(avg_aqi_comparison)
```

Which months consistently show the worst air quality across Indian states —
(Consider top 10 states with high distinct areas)



```python



aqi['date'] = pd.to_datetime(aqi['date'], errors='coerce', dayfirst=True)

aqi['month'] = aqi['date'].dt.month


top_states = aqi.groupby('state')['area'].nunique().sort_values(ascending=False).head(10).index.tolist()


filtered_df = aqi[aqi['state'].isin(top_states)]


monthly_avg = filtered_df.groupby(['state', 'month'])['aqi_value'].mean().reset_index()


worst_months = monthly_avg.loc[monthly_avg.groupby('state')['aqi_value'].idxmax()]


print(worst_months.sort_values(by='aqi_value', ascending=False))
```

List the top two most reported disease illnesses in each state over the past three
years, along with the corresponding average Air Quality Index (AQI) for that
period.

```python
import pandas as pd
idsp['outbreak_starting_date'] = pd.to_datetime(idsp['outbreak_starting_date'], dayfirst=True)
aqi['date'] = pd.to_datetime(aqi['date'], dayfirst=True)


latest_date = max(idsp['outbreak_starting_date'].max(), aqi['date'].max())
three_years_ago = latest_date - pd.DateOffset(years=3)


disease_recent = idsp[idsp['outbreak_starting_date'] >= three_years_ago]
aqi_recent = aqi[aqi['date'] >= three_years_ago]


disease_summary = (disease_recent.groupby(['state', 'disease_illness_name'])['cases']
                   .sum().reset_index()
                   .sort_values(['state', 'cases'], ascending=[True, False]))

top2_diseases = disease_summary.groupby('state').head(2)

avg_aqi = aqi_recent.groupby('state')['aqi_value'].mean().reset_index()
avg_aqi.rename(columns={'aqi_value': 'avg_aqi'}, inplace=True)


final_df = pd.merge(top2_diseases, avg_aqi, on='state', how='left')
final_df.head(76)

##print(final_df.sort_values(['state', 'cases'], ascending=[True, False]))
```

List the top 5 states with high EV adoption and analyse if their average AQI is
significantly better compared to states with lower EV adoption

```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
#ev_data = vahan[vahan['fuel'].str.upper() == 'ELECTRIC']
ev_state_counts = vahan.groupby('state')['value'].sum().reset_index()
ev_state_counts = ev_state_counts.sort_values(by='value', ascending=False)
ev_state_counts
```

```python
top_5_states = ev_state_counts.head(5)['state'].tolist()
print("Top 5 EV adoption states:", top_5_states)
aqi['aqi_value'] = pd.to_numeric(aqi['aqi_value'], errors='coerce')
avg_aqi = aqi.groupby('state')['aqi_value'].mean().reset_index()
```

```python
group_A = avg_aqi[avg_aqi['state'].isin(top_5_states)]['aqi_value'].dropna()
group_B = avg_aqi[~avg_aqi['state'].isin(top_5_states)]['aqi_value'].dropna()
```

```python
t_stat, p_value = ttest_ind(group_A, group_B, equal_var=False)
```

```python
print("Average AQI - Top 5 EV States:\n", group_A.mean())
print("Average AQI - Other States:\n", group_B.mean())
print(f"T-test p-value: {p_value}")

if p_value < 0.05:
    print("✅ The difference in AQI is statistically significant.")
else:
    print("❌ No significant difference in AQI between high and low EV adoption state")
```
